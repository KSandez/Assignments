# Purchasing Analysis (Total)

```python
import pandas as pd

purchasefile = "Resources/purchase_data_3.csv"
```

```python
### Purchasing Analysis (Total)
purchase_df = pd.read_csv(purchasefile, encoding="ISO-8859-1")
purchase_df.head()
```

```python
uniqueitems = len(purchase_df["Item ID"].unique())
#print("Number of Unique Items: " + str(uniqueitems_pd))

avg_price = "${:.2f}".format(purchase_df["Price"].mean())
#print("Average Price: " + "${:.2f}".format(avg_price))

#print("Total Number of Purchases: " + str(len(purchase_df)))
total_purchases = len(purchase_df)

#print("Total Revenue: " + "${:.2f}".format(purchase_df["Price"].sum()))

total_revenue = "${:.2f}".format(purchase_df["Price"].sum())

purchase_table = pd.DataFrame({"Number of Unique Items": [uniqueitems],
                               "Average Price": [avg_price],
                               "Total Number of Purchases": [total_purchases],
                               "Total Revenue": [total_revenue]
                              })
purchase_table
```

# Age Demographics

```python
import pandas as pd
import numpy as np

playerfile = "Resources/players_complete.csv"

players_df = pd.read_csv(playerfile, encoding="ISO-8859-1")
players_df.head(10)
```

```python
### Player Count
count = players_df["Player ID"].count()

count_table = pd.DataFrame({"Total Players": [count]})
count_table
```

```python
bins = [0, 10, 14, 19, 24, 29, 34, 39, 100]
age_group = ["under 10", "10-14", "15-19", "20-24","25-29","30-34","35-39","40 & up"]

df["Age Group"] = pd.cut(players_df["Age"], bins, labels=age_group)
df.head(10)
```

```python
age_groupcnt = df.groupby("Age Group")
age_groupcnt
```

```python
age_cnt = pd.DataFrame(age_groupcnt["Age Group"].count())
age_cnt
```

```python
age_cnt = age_cnt.rename(columns={"Age Group": "Total Count"})
age_cnt
```

```python
age_cnt = age_cnt.reset_index()
age_cnt
```

```python
percentage_cnt = pd.DataFrame(((age_groupcnt["Age Group"].count())/count)*100)
percentage_cnt
```

```python
percentage_cnt = percentage_cnt.rename(columns={"Age Group": "Percentage of Players"})
percentage_cnt
```

```python
percentage_cnt = percentage_cnt.reset_index()
percentage_cnt
```

```python
percentage_cnt["Percentage of Players"] = percentage_cnt["Percentage of Players"].map("{:.2f}%".format)
percentage_cnt
```

```python
age_summary = percentage_cnt.merge(age_cnt, on="Age Group")
age_summary
```

# Top 5 Spenders

```python
import pandas as pd
import numpy as np
purchasefile = "Resources/purchase_data_3.csv"
purchase_df = pd.read_csv(purchasefile, encoding="ISO-8859-1")
purchase_df.head()
```

```python
grouped_SN = purchase_df.groupby("SN")
```

```python
purchase_total_by_SN = pd.DataFrame(grouped_SN["Price"].sum().map("${:.2f}".format))
purchase_total_by_SN.head(7)
```

```python
purchase_total_by_SN = purchase_total_by_SN.rename(columns={"Price":"Total Purchase Value"})
purchase_total_by_SN = purchase_total_by_SN.reset_index()
purchase_total_by_SN.head(10)
```

```python
purchase_count_by_SN = pd.DataFrame(grouped_SN["Item ID"].count())
purchase_count_by_SN.head(15)
```

```python
purchase_count_by_SN = purchase_count_by_SN.rename(columns={"Item ID":"Purchase Count"})
purchase_count_by_SN = purchase_count_by_SN.reset_index()
purchase_count_by_SN.head(5)
```

```python
purchase_avg_by_SN = pd.DataFrame(grouped_SN["Price"].mean().map("${:.2f}".format))
purchase_avg_by_SN.head(7)
```

```python
purchase_avg_by_SN = purchase_avg_by_SN.rename(columns={"Price":"Average Purchase Price"})
purchase_avg_by_SN = purchase_avg_by_SN.reset_index()
purchase_avg_by_SN.head(5)
```

```python
mergr_purccnt_avgpurc = purchase_count_by_SN.merge(purchase_avg_by_SN, on = "SN")
mergr_purccnt_avgpurc.head(8)
```

```python
final_merge = mergr_purccnt_avgpurc.merge(purchase_total_by_SN, on = "SN")
final_merge.head(9)
```

```python
top_spender_df = final_merge.sort_values("Total Purchase Value", ascending=False)
top_spender_df.head(5)
```

```python
top_spender_df = top_spender_df.reset_index()
top_spender_df.head(5)
```

# Most Popular Items

```python
import pandas as pd
import numpy as np

purchasefile = "Resources/purchase_data_3.csv"
purchase_df = pd.read_csv(purchasefile, encoding="ISO-8859-1")
purchase_df.head()
```

```python
grouped_item = purchase_df.groupby(['Item ID','Item Name'])
```

```python
pop_item = pd.DataFrame(grouped_item['Price'].sum())
pop_item.head(8)
```

```python
pop_item = pop_item.rename(columns={"Price":"Total Purchase Value"})
pop_item.head(8)
```

```python
pop_item = pop_item.reset_index()
pop_item.head(8)
```

```python
pop_item = pop_item.sort_values("Total Purchase Value", ascending=False)
pop_item.head(8)
```

```python
grouped_item2 = purchase_df.groupby(['Item ID','Price'])
```

```python
itemcnt = pd.DataFrame(grouped_item2['Item Name'].count())
itemcnt.head(8)
```

```python
itemcnt = itemcnt.rename(columns={"Item Name":"Purchase Count"})
itemcnt.head(8)
```

```python
itemcnt = itemcnt.sort_values("Purchase Count", ascending=False)
itemcnt.head(8)
```

```python
itemcnt = itemcnt.reset_index()
itemcnt.head(8)
```

```python
merge_pop = itemcnt.merge(pop_item, on = "Item ID")
merge_pop.head(9)
```

```python
most_popular = merge_pop.sort_values(["Purchase Count","Total Purchase Value"], ascending=False)
most_popular.head(9)
```

```python
#most_popular = most_popular.reset_index()
most_popular.head(9)
```

```python
clean_most_popular = most_popular[["Item ID","Item Name","Purchase Count","Price","Total Purchase Value"]]
clean_most_popular.head(9)
```

```python
clean_most_popular["Price"] = most_popular["Price"].map("${:.2f}".format)
clean_most_popular["Total Purchase Value"] = most_popular["Total Purchase Value"].map("${:.2f}".format)
clean_most_popular.head(10)
```

# Most Profitable Items

```python
most_profits = most_popular.sort_values(["Total Purchase Value","Purchase Count"],ascending=False)
most_profits = most_profits[["Item ID","Item Name","Purchase Count","Price","Total Purchase Value"]]
most_profits.head(10)
```
